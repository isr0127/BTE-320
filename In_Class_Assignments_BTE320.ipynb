**Income Statement using Simple Income Statement Calculation:**

```python
# Income Statement Program

# Get sales input from user
sales = float(input("Enter sales ($): "))

# Calculate Cost of Goods Sold (40% of Sales)
cost_of_goods_sold = sales * 0.40

# Calculate Gross Profit
gross_profit = sales - cost_of_goods_sold

# Get tax rate from user
tax_rate = float(input("Enter tax rate (%): "))

# Calculate Taxes and Net Income
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

# Display the Income Statement
print("\n Income Statement ")
print(f"Sales:              ${sales:,.2f}")
print(f"Cost of Goods Sold: ${cost_of_goods_sold:,.2f}")
print(f"Gross Profit:       ${gross_profit:,.2f}")
print(f"Taxes:              ${taxes:,.2f}")
print(f"Net Income:         ${net_income:,.2f}")

```

**Numerical Calculator (if-elif-else statements):**

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
op = input("choose numerical operator: ")

if op == "+":
    res = num1 + num2
elif op == "-":
    res = num1 - num2
elif op == "*":
    res = num1 * num2
elif op == "/":
    res = num1 / num2
elif op == "//":
    res = num1 // num2
elif op == "**":
    res = num1 ** num2
elif op == "%":
    res = num1 % num2
else:
  print("Invalid operation")
  res = None

print(f"Result: {res}")

```

**Parking garage (for/while loop):**

```python
for hours in range (1,9):
  fee = 5 + (2.5 * hours)

  if fee < 10 :
    fee = 10.0
  elif fee > 20 :
    fee = 20.0

  print(hours, fee)
```

**Movie Theater (for/while loop):**

```python
import math

print("Ad$ Profit")

for advertising_dollars in range(0, 201,25):
  additional_attendees= 2 * math.sqrt(advertising_dollars)
  total_attendees = 20 + additional_attendees
  revenue = total_attendees * 10
  profit = revenue - advertising_dollars - 200

  print(float(advertising_dollars)), round(profit, 1)
```

**Rock-Paper-Scissors**

```python
import random

play = input("Do you want to play? (yes/no): "). lower()

while play == "yes":
  user = input("Choose rock, paper, or scissors: ").lower()
  computer = random.choice(["rock", "paper", "scissors"])

  print("Computer chose:", computer)

  if user == computer:
    print("It's a tie!")
  elif user == "rock" and computer == "scissors":
    print("You Win!")
  elif user == "paper" and computer == "rock":
    print("You Win!")
  elif user == "scissors" and computer == "paper":
    print("You Win!")
  elif user in ["rock", "paper", "scissors"] :
    print("You Lose!")
  else:
    print("Invalid choice")

  play = input("Do you want to play again? (yes/no): "). lower()
```

**Append items to list iteratively (while loop):**

```python
numbers = []

num = float(input("Enter a number (0 to stop): "))

while num != 0:
  numbers.append(num)
  num = float(input("Enter a number (0 to stop): "))

  numbers.sort(reverse=True)
  print(numbers)
```

**Dynamic dictionary; N distinct key: value pairs (while loop/ if-elif-else statements):**

```python
counts = {}

N = int(input("Enter how many names you want to input: "))

for i in range(N):
  key = input("Enter a name: ")

  if key in counts:
    counts[key] += 1
  else:
    counts[key] = 1

print(counts)
```

**Recursive function (string reversal):**

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    #retun s[-1] + reverse(s[:-1]) # Take the last char and put it first
    #return reverse (s[1:]) + s[0] # Take the first char and put it last
    return s[-1] + reverse(s[1:-1]) + s[0]

print(reverse('12345'))
print(reverse('hello'))
```
