**Income Statement using Simple Income Statement Calculation:**

```python
# Income Statement Program

# Get sales input from user
sales = float(input("Enter sales ($): "))

# Calculate Cost of Goods Sold (40% of Sales)
cost_of_goods_sold = sales * 0.40

# Calculate Gross Profit
gross_profit = sales - cost_of_goods_sold

# Get tax rate from user
tax_rate = float(input("Enter tax rate (%): "))

# Calculate Taxes and Net Income
taxes = gross_profit * (tax_rate / 100)
net_income = gross_profit - taxes

# Display the Income Statement
print("\n Income Statement ")
print(f"Sales:              ${sales:,.2f}")
print(f"Cost of Goods Sold: ${cost_of_goods_sold:,.2f}")
print(f"Gross Profit:       ${gross_profit:,.2f}")
print(f"Taxes:              ${taxes:,.2f}")
print(f"Net Income:         ${net_income:,.2f}")

```

**Rock-Paper-Scissors**

```python
import random

play = input("Do you want to play? (yes/no): "). lower()

while play == "yes":
  user = input("Choose rock, paper, or scissors: ").lower()
  computer = random.choice(["rock", "paper", "scissors"])

  print("Computer chose:", computer)

  if user == computer:
    print("It's a tie!")
  elif user == "rock" and computer == "scissors":
    print("You Win!")
  elif user == "paper" and computer == "rock":
    print("You Win!")
  elif user == "scissors" and computer == "paper":
    print("You Win!")
  elif user in ["rock", "paper", "scissors"] :
    print("You Lose!")
  else:
    print("Invalid choice")

  play = input("Do you want to play again? (yes/no): "). lower()
```

**Recursive function (string reversal):**

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    #retun s[-1] + reverse(s[:-1]) # Take the last char and put it first
    #return reverse (s[1:]) + s[0] # Take the first char and put it last
    return s[-1] + reverse(s[1:-1]) + s[0]

print(reverse('12345'))
print(reverse('hello'))
```

**ROI Calculator**

```python
def roi(cost, revenue):
  #check numeric types
  if not isinstance(cost, (int,float)) or not isinstance(revenue, (int, float)):
    raise TypeError("Both cost and revenue must be numbmers")

  #check cost > 0
  if cost <= 0:
    raise ValueError("Cost must be greater than 0")

  #calculate ROI
  result = (revenue - cost) / cost

  return result


print(roi(100,120))
```

**Loan Payment Calculator:**

```python
def loan_payment(principal, annual_rate, years):
  #validate inputs
  if not all(isinstance(x, (int, float)) for x in [principal, annual_rate, years]):
    raise TypeError("All inputs must be numeric")

  if principal <= 0 or annual_rate <= 0 or years <= 0:
    raise ValueError("All inputs must be positive")

  #convert values
  r = annual_rate / 12
  n = years * 12

  #calculate payment
  payment = (r * principal) / (1-(1*r)**(-n))

  return payment

print(loan_payment(1000, 0.05,5))
```

**Quality Control Managaer (exceptions):**

```python
def getValues(n):
  values = []

  for i in range(n):
    while True:
      try:
        x = float(input(f"Enter value {i+1}: "))

        if x < 0:
          raise ValueError

        values.append(x)
        break

      except ValueError:
        print("Invalid input. Please enter a positive number.")

  return values


def calcAverage(alist):
  if len(alist) <= 3:
    print("Not enough values to caculate average after excluding first 3.")
    return 0


  new_list = alist[3:]
  avg = sum(new_list) / len(new_list)
  return avg

try:
  n = int(input("Enter number of values: "))

  if n <= 0:
    raise ValueError

  data = getValues(n)
  result = calcAverage(data)

  print("Average(excluding first 3 values) =", result)

except ValueError:
  print("Invalid input. Please enter a positive integer.")
```

**Greeting Class:**

```python
class Greeting:

  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return f"{self.first} {self.last}"

person1 = Greeting("Isobel", "Reaves")

person1.display(False)

person1.display(True)

print(person1)
```

**Student Class:**

```python
class Student:

  def __init__(self, name, number, courses):
    self.name = name
    self.number = number
    self.courses = courses

  def get_names(self):
    return self.name

  def get_courses(self):
    return self.courses

  def set_name(self, new_name):
    self.name = new_name

  def __len__(self):
    return len(self.courses)

  def enroll(self, course):
    if course not in self.courses:
      self.courses.append(course)
    else:
      print("Course already enrolled.")


student1 = Student("Isobel Reaves", 12345, ["Math", "English"])

print(student1.get_names())
print(student1.get_courses())

student1.set_name("Isobel R.")
print(student1.get_courses())

student1.enroll("Science")
student1.enroll("Math")

print(student1.get_courses())

print("Number of courses:", len(student1))
```

**Class Point and Rectangle:**

```python
import math

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self,other):
    return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

class Rectangle:

  def __init__(self, p1, p2, p3, p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4

  def width(self):
    return self.p1.distance(self.p2)

  def height(self):
    return self.p2.distance(self.p3)

  def area(self):
    return self.width() * self.height()

p1 = Point (0,0)
p2 = Point (4,0)
p3 = Point (4,3)
p4 = Point (0,3)

rect = Rectangle(p1, p2, p3, p4)

print("Width:", rect.width())
print("Height:", rect.height())
print("Area:", rect.area())
```

**Class Employee- Manager**

```python
class Employee:
  def __init__(self, name, hourly_wage, hours):
    self.name = name
    self.__hourly_wage = hourly_wage
    self.hours = hours

  @property
  def wage(self):
    return self.__hourly_wage

  @wage.setter
  def wage(self, newWage):
    assert newWage > 0
    self.__hourly_wage = newWage


  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employees name: {self.name}\nHourly charge: {self.__hourly_wage}$"

class Manager(Employee):
  def __init__(self, name, hourly_wage, hours, bonus):
    super().__init__(name, hourly_wage, hours)
    self.bonus = bonus #float between 0 - 1; e.g., 0.2 means 20% extra

  def calculate_pay(self):
    return (1 + self.bonus) * self.wage * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary: {employee.calculate_pay()}$.")
print(f"Manager {manager.name} weekly salary: {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```
